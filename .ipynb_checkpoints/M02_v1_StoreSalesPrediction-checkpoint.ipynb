### 0. Imports

```python
import pandas as pd
import inflection as ifn
from IPython.display import display
import numpy as np
import math
import seaborn as sns
from matplotlib import pyplot as plt
```

    0.1 Funções auxiliáres

    0.2 Carga de dados

```python
data_sales = pd.read_csv('Data/train.csv', low_memory=False)
data_store = pd.read_csv('Data/store.csv', low_memory=False)

#pre análise dos arquivos
display(data_sales.head(2), data_store.head(2))
```

```python
#Concatenado DF's
data = pd.merge(data_sales,data_store, how='left', on='Store')
```

### 1. Descrição dos Dados

```python
#Bkp da base de dados
df = data.copy()
```

```python
display(data.head(3))
```

    1.1. Renomeando colunas

```python
old_cols = df.columns

###Criando função para alterar os nomes das colunas
snakecase = lambda x : ifn.underscore(x)
##Alteranddo os dados e criando lista de novos nomes
new_cols = list(map(snakecase, old_cols))

#Renomeando colunas
df.columns = new_cols
```

```python
old_cols
```

    1.2. Dimensionamento dos Dados

```python
display("O conjunto de dados é composto por {} linhas e {} colunas".format(df.shape[0], df.shape[1]))
```

    1.3. Análise dos tipos dos Dados

```python
display(df.dtypes)
df.date = pd.to_datetime(df.date)
```

    1.4. Análise de valores 'NA'

```python
display(df.isna().sum())
```

    1.5. Substituindo valores 'NA'

```python
### 1. competition_distance         
##Em análise a descrição da coluna, os dados serão substituídos por valores extremamente altos
display("O maior valor originalmente contidoem 'competition_distance' era: {}".format(df.competition_distance.max()))
df.competition_distance = df.competition_distance.fillna((df.competition_distance.max() * 2))
```

```python
### 2. competition_open_since_month  
df.competition_open_since_month = df.apply(lambda x : x['date'].month if math.isnan(x['competition_open_since_month'])  else x['competition_open_since_month'], axis=1 )       
df.competition_open_since_month.isna().sum()        
```

```python
### 3. competition_open_since_year     
df.competition_open_since_year = df.apply(lambda x : x['date'].year if math.isnan(x['competition_open_since_year'])  else x['competition_open_since_year'], axis=1 ) 
df.competition_open_since_year.isna().sum()        
```

```python
### 4. promo2_since_week        
df.promo2_since_week = df.apply(lambda x : x['date'].week if math.isnan(x['promo2_since_week'])  else x['promo2_since_week'], axis=1 ) 
df.promo2_since_week.isna().sum()
```

```python
###promo2_since_year   
df.promo2_since_year = df.apply(lambda x : x['date'].year if (math.isnan(x['promo2_since_year']))  else x['promo2_since_year'], axis=1 ) 
df.promo2_since_year.isna().sum()            
```

```python
#promo_interval
month_map = {1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun', 7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'}

df.fillna(0, inplace=True)

df['month_map'] = df['date'].dt.month.map( month_map)

df['is_promo'] =  df[['month_map', 'promo_interval']].apply(lambda x : 0 if x['promo_interval'] == 0 else 1 if x['month_map'] in x['promo_interval'].split(',') else 0, axis=1)
```

```python
df.dtypes
```

```python
df.competition_open_since_month = df.competition_open_since_month.astype(int)
df.competition_open_since_year = df.competition_open_since_year.astype(int)
df.promo2_since_week = df.promo2_since_week.astype(int)
df.promo2_since_year = df.promo2_since_year.astype(int)
```

    1.7. Statística descritiva

- Tendência central = média - mediana
- Dispersão = std - min - max - range - skew - kurtosis

```python
cat_atributes = df.select_dtypes(exclude = ['int64','float64','datetime64[ns]'])
n_atrib = df.select_dtypes(include = ['int64','float64'])
```

        1.7.1. Atributos numéricos  

```python
num_atributes = df.describe().transpose()
num_atributes
```

```python
num_atributes.columns
```

```python
num_atributes['range'] = pd.DataFrame(n_atrib.apply(lambda x : x.max() - x.min()))
num_atributes['skew'] = pd.DataFrame(n_atrib.apply(lambda x : x.skew()))
num_atributes['kurtosis'] = pd.DataFrame(n_atrib.apply(lambda x : x.kurtosis()))
num_atributes = num_atributes.reindex(columns=['min', 'max', 'range','25%', '50%', '75%', 'std', 'kurtosis'])
num_atributes
```

### 2.0 Helper Functions

    * Cramer Value

```python
def cramer_v(x,y):    
    # Criação da tabela de contingência (Confusion Matrix 'cm')
    cm = pd.crosstab(x,y)
    cm = np.asmatrix(cm)

    #Coletando valores de n(soma de todos os valores)
    n = cm.sum()

    # r (linhas) e k(colunas)
    r,k = cm.shape

    #Cálculo de chi2, onde será utilizado apenas o primeiro valor (valor do teste)
    chi2 = ss.chi2_contingency(cm)[0]
    
    #Correções sugeridas para que não haja tendência de alta na fórmula
    chi2corr = max(0 , chi2 - (k-1) * (r-1) / (n-1))
    kcorr = k - (k-1)**2/(n-1)
    rcorr = r - (r-1)**2/(n-1)
    
    return np.sqrt( (chi2corr/n) / ( min(kcorr-1,rcorr-1) ) )
```

    * Mape and ml_error

```python
def mape_ (y,yhat):
    return np.mean(np.abs( (y-yhat) /y ))

def ml_error(model_name, y, yhat):
    mae = mean_absolute_error(y,yhat)
    mape = mean_absolute_percentage_error(y,yhat)
    mape1 = mape_(y,yhat)
    rmse = np.sqrt(mean_squared_error(y,yhat))
    
    return pd.DataFrame({
        'Model Name': model_name,
        'MAE': mae,
        'MAPE' : mape,
        'MAPE1' : mape1,
        'RMSE' : rmse
    }, index=[0])
```

    * Cross-Validation

```python
def ts_cross_validation(model_name,model, x_training, number_of_weeks, kfold):
    mae  = []
    mape = []
    rmse = []
    for k in reversed(range(1,kfold + 1)):
        validation_start_date = x_training['date'].max() -datetime.timedelta(days= k * number_of_weeks * 7)
        validation_end_date = x_training['date'].max() -datetime.timedelta(days= (k - 1) * number_of_weeks * 7)
    
        # Filtering dataset
        training = x_training[x_training['date'] < validation_start_date]
        validation = x_training[(x_training['date'] >= validation_start_date) & (x_training['date'] <= validation_end_date)]
    
        # Training prepare
        xtraining = training.drop(columns=['date', 'sales'])
        ytraining = training['sales']
    
    
        # Validation prepare
        xvalidation = validation.drop(columns=['date', 'sales'])
        yvalidation = validation['sales']
    
        #Training the model
    
        #Model
        m = model.fit(xtraining, ytraining)
    
        #Prediction
        yhat = m.predict(xvalidation)
    
        #Performance
        m_result = ml_error('Linear Regression', np.expm1(yvalidation), np.expm1(yhat))
    
        # store performance of each interation
        mae.append(m_result['MAE'])
        mape.append(m_result['MAPE'])
        rmse.append(m_result['RMSE'])
    
    #Cross_Validation result
    
    result = pd.DataFrame(
        { 'Model Name': model_name,
          'MAE CV' : np.round( np.mean(mae), 2 ).astype(str) + ' +/- ' + np.round(np.std(mae)).astype(str),
          'MAPE CV' : np.round( np.mean(mape), 2 ).astype(str) + ' +/- ' + np.round(np.std(mape)).astype(str),
          'RMSE': np.round( np.mean(rmse), 2 ).astype(str) + ' +/- ' + np.round(np.std(rmse)).astype(str)}, index = [0]
    )

    return result     
```

### 7.0. Machine Learning Modelling

#### Data prepare

```python
cols_selected_boruta = ['store', 'promo', 'store_type', 'assortment', 'competition_distance', 'competition_open_since_month', 'competition_open_since_year', 'promo2', 'promo2_since_week', 'promo2_since_year', 'competition_time_month',
     'promo_time_week', 'day_of_week_sin', 'day_of_week_cos', 'month_sin', 'month_cos', 'day_sin', 'day_cos', 'weekofyear_sin', 'weekofyear_cos']

cols_boruta_full = ['store', 'promo', 'store_type', 'assortment', 'competition_distance', 'competition_open_since_month', 'competition_open_since_year', 'promo2', 'promo2_since_week', 'promo2_since_year', 'competition_time_month',
     'promo_time_week', 'day_of_week_sin', 'day_of_week_cos', 'month_sin', 'month_cos', 'day_sin', 'day_cos', 'weekofyear_sin', 'weekofyear_cos','date']
```

```python
#Colecting data for model training
x_train = X_train[cols_selected_boruta]
x_test = X_test[cols_selected_boruta]
```

#### 7.1 Linear Models

##### 7.1.1 Average Model

```python
aux = x_test.copy()
aux['sales'] = Y_test.copy()

#Predictions
aux2 = aux[['store','sales']].groupby('store').mean().reset_index().rename(columns={'sales':'predictions'})
aux = pd.merge(aux, aux2, how='left', on='store')
y_hat_baseline =aux['predictions']

#Performance
average_result = ml_error('Average Model',np.expm1( Y_test), np.expm1(y_hat_baseline))
```

##### 7.1.2 Linear Regression Model

```python
#Model
lr = LinearRegression().fit(x_train, Y_train)

#Prediction
yhat_lr = lr.predict(x_test)

#Performance
lr_result = ml_error('Linear Regression', np.expm1(Y_test), np.expm1(yhat_lr))

```

##### 7.1.3 Linear Regression Regularized Model

```python
#Model
lrr = Lasso(alpha=0.0001).fit(x_train, Y_train)

#Prediction
yhat_lrr = lrr.predict(x_test)

#Performance
lrr_result = ml_error('LR Regularized', np.expm1(Y_test), np.expm1(yhat_lrr))
```

##### Result linear models

```python
result_linear = pd.concat([average_result, lr_result, lrr_result]).reset_index(drop=True)
result_linear
```

#### 7.2 Regression Models

##### 7.2.1 Rando Forest Regression

```python
#Model
rfr = RandomForestRegressor(n_estimators=100, n_jobs=-1, random_state=42).fit(x_train, Y_train)

#Prediction
yhat_rfr = rfr.predict(x_test)

#Performance
rfr_result = ml_error('Random Forest Regression', np.expm1(Y_test), np.expm1(yhat_rfr))
```

##### 7.2.2 XGBoost Regression

```python
#Model
xgb_model = xgb.XGBRFRegressor(
    n_jobs=-1,
    n_estimators=100,
).fit(x_train, Y_train)

#Prediction
yhat_xgb = xgb_model.predict(x_test)

#Performance
xgb_result = ml_error('XGBoost Regressior', np.expm1(Y_test), np.expm1(yhat_xgb))
```

##### Result regression models

```python
result_regression = pd.concat([rfr_result,xgb_result]).reset_index(drop=True)
result_regression
```

#### 7.3 Compare Model's Preformance

```python
result = pd.concat([result_linear, result_regression]).reset_index(drop=True).sort_values('RMSE', ascending=True)
result
```

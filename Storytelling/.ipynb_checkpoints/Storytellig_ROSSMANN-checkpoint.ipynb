- Reunião mensal de resultados;
- CFO solicitou aos seus gerentes um relatório de previsão de vendas das próximas 6 semanas de cada loja;

## <font color='blue'> 2. Desafio </font>

### Necessidade
- Relatório de previsão de vendas que sirva como base para definição do Budget de reforma das Lojas;

### Causas
- Altos níveis de divergência na predição de vendas atual;
- Processo de previsões de vendas atual totalmente empírico e baseado em experiências passadas;
- Previsões feitas de forma manual em cada uma das 1.115 lojas da ROSSMANN;
- Modelo contra intuitívo, extenso e que dificultava o acesso por meios remotos;

## <font color='blue'> 3. Proposta de Solução </font>

### Técnicas e ferramentas
- Por meio de ferramentas de Machine Learning, criar modelo de Previsão de Vendas de todas as lojas nas próximas 6 semanas; 
- Viabilizar o acesso das informações retornadas pelo modelo via smartphone;

```python
from IPython.display import Image
import pandas as pd
```

```python
###### Deletar
sales = pd.read_csv('Data/train.csv', low_memory=False)
store = pd.read_csv('Data/store.csv', low_memory=False)

df = pd.merge(sales, store, how='left', on='Store')
```

```python
num_atrib = df.select_dtypes(include = ['int64','float64'])
num_atributes = df.describe().transpose()


num_atributes['range'] = pd.DataFrame(num_atrib.apply(lambda x : x.max() - x.min()))
num_atributes['kurtosis'] = pd.DataFrame(num_atrib.apply(lambda x : x.kurtosis()))
num_atributes = num_atributes.reindex(columns=['range', 'min', 'max', '25%', '50%', '75%', 'std', 'kurtosis'])
```
